## **Series**

```python
# Creating series from NumPy ndarray with customize index

import numpy as np
import pandas as pd

# Step 1: Create a NumPy array with daily sales numbers
daily_sales_array = np.array([250, 300, 400, 500, 450])

# Step 2: Create a Series from the NumPy array
daily_sales = pd.Series(daily_sales_array, index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])

# Step 3: Print the Series
print("Daily Sales Data:")
print(daily_sales)
```

```python
#CREATING SERIES FRON DICTIONARY

# Dictionary data
data = {'a': 10, 'b': 20, 'c': 30, 'd': 40}

# Create a Pandas Series
series = pd.Series(data)

# Display the Series
print("Series created from dictionary:")
print(series)
```

```python
cloude_cost = pd.Series ([12000, 34533,76463, 34373, 87432], index=['jan', 'fab', 'mar', 'apr', 'may'])
high_cost = cloude_cost >13000
print("\n",high_cost)

outlire_month = (cloude_cost<12000) | (cloude_cost > 34573)
print("\n",outlire_month)

within_budget = ~high_cost
print("\n",within_budget)

filtered_data = cloude_cost[high_cost]
print("\n",filtered_data)

where_example = cloude_cost.where(cloude_cost<13000)
print("\n",where_example)
```

```python
# Creating logistics shipment DataFrame
df = pd.DataFrame({
    'ShipmentID': [1001, 1002, 1003, 1004, 1005],
    'Weight_kg': [12.5, 8.0, 15.0, 7.5, 10.0],
    'Status': ['Delivered', 'In Transit', 'Delayed', 'Delivered', 'Delayed'],
    'Fragile': [True, False, True, False, True],
    'Shipping_Type': ['Air', 'Ground', 'Air', 'Ground', 'Air']
})

#get all critical shifment
critical_df = df[df['Status'] == 'Critical']
print(critical_df)

#get all heavy and air-shipped packeges
heavy_air_packages = df[(df['Weight_kg'] > 10) & (df['Shipping_Type'] == 'Air')]
print(heavy_air_packages)
```
